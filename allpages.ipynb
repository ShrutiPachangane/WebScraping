```python
url='https://www.themoviedb.org/movie'
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
html_data = requests.get(url).text
print(html_data)
```

```python
soup_data = BeautifulSoup(html_data,'lxml')
```

```python
url_lst=[]
base_url='https://www.themoviedb.org/movie?page='
for item in range(1,51):
    url_lst.append(base_url+str(item))
    
for val in url_lst:
    print(val)
```

```python
dir_common_url='https://www.themoviedb.org'
common_url='https://www.themoviedb.org'
h2_data=soup_data.find_all('div',class_='card style_1')
page_1_lst = []
for url_name in url_lst:
    html_data = requests.get(url_name).text
    soup_data = BeautifulSoup(html_data,'lxml')
    h2_data=soup_data.find_all('div',class_='card style_1')
    for item in h2_data:
        movie_url=item.find('a')['href']
        movie_url=common_url+movie_url
        #print(movie_url)
        movie_data=requests.get(movie_url).text
        #print(movie_data)
        soup_data1=BeautifulSoup(movie_data,'lxml')
        #soup_data1
        title=soup_data1.find_all('section',class_='header poster')
        #title
        for val in title:
            title_name=val.find('a').text
            rating=val.find('div',class_='user_score_chart')['data-percent']
            #genre=val.find('span',class_='genres').text.replace('\n','').replace('\xa0','')
            try:
                genre=val.find('span',class_='genres').text.replace('\n','').replace('\xa0','')
            except:
                 genre = 'Not Available'
            #release_date=val.find('span',class_='release').text.strip().replace('(IN)','').strip()
            try:
                release_date=val.find('span',class_='release').text.strip().replace(" ","")[0:10].replace("-","/")
            except:
                 release_date = 'Not Available'
            #runtime=val.find('span',class_='runtime').text.strip()
            try:
                runtime=val.find('span',class_='runtime').text.strip()
            except:
                runtime='Not Available'
            #director=val.find_all('li',class_='profile')[0]('p')[0].text
            try:
                director=val.find_all('li',class_='profile')[0]('p')[0].text
            except:
                director='Not Available'
            try:
                director_url=val.find_all('li',class_='profile')[0]('p')[0]('a')[0]['href']
                director_url=dir_common_url+director_url
            except:
                director_url='Not Available'
            #director_url=val.find_all('li',class_='profile')[0]('p')[0]('a')[0]['href']
            #director_url=dir_common_url+director_url
    
            all_movie_info={
                'title_name':title_name,
                'rating':rating,
                'genre':genre,
                'release_date':release_date,
                'runtime':runtime,
                'director':director,
                'director_url':director_url
                }
            page_1_lst.append(all_movie_info)
print(page_1_lst)
   
    
```
